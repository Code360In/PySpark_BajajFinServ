# 01-pyspark-string-date.py

```
# 01-pyspark-string-date.py
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

# Create SparkSession
spark = SparkSession.builder \
               .appName('PySparkExamples') \
               .getOrCreate()

from pyspark.sql.functions import *

df = spark.createDataFrame([["02-03-2013"], ["05-06-2023"]], ["input"])
df.printSchema()  # here date is in string type
df.show()
```

```
df1 = df.select(col("input"), to_date(col("input"), "MM-dd-yyyy").alias("date"))
df1.printSchema()
df1.show()
```

```
# SQL
df2 = spark.sql("select to_date('02-03-2013', 'MM-dd-yyyy') date")
df2.printSchema()
df2.show()
```
