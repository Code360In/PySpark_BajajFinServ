## 03-pyspark-rdd-reduceByKey.py

```
# 03-pyspark-rdd-reduceByKey.py 
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySparkExamples').getOrCreate()

data = [('Mango', 1), ('Banana', 1), ('Mango', 1), ('Apple', 1), ('Apple', 1), ('Apple', 1),
        ('Banana', 1), ('Guava', 1), ('Banana', 1), ('Mango', 1), ('Pineapple', 1), ('Pineapple', 1),
        ('Apple', 1)]

rdd = spark.sparkContext.parallelize(data)

rdd2 = rdd.reduceByKey(lambda a, b: a + b)
print (rdd2.collect())
for element in rdd2.collect():
    print(element)
```
