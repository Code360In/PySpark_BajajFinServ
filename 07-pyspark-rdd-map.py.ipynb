## 07-pyspark-rdd-map.py

```
# 07-pyspark-rdd-map.py
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('PySparkExamples').getOrCreate()

data = ["Apple", "Pineapple", "Guava", "Carrot", "Lemon", "Banana", "Apple", "Pineapple",
        "Carrot", "Lemon", "Banana", "Apple", "Pineapple"]

rdd = spark.sparkContext.parallelize(data)

rdd2 = rdd.map(lambda x: (x,1))
print (rdd2.collect())
for element in rdd2.collect():
    print(element)
```

```
data = [('Kamal', 'Dasgupta', 'M', 30),
        ('Sanjib', 'Das', 'F', 41),
        ('Bidhun', 'Biswas', 'M', 62)]

columns = ["Firstname", "Lastname", "Gender", "Salary"]
df = spark.createDataFrame(data = data, schema = columns)
df.show()
```

```
rdd2 = df.rdd.map(lambda x: (x[0] + ", " + x[1], x[2], x[3] * 2))  
df2 = rdd2.toDF(["Name", "Gender", "New_Salary"])
df2.show()
```

```
#Referring Column Names
rdd2 = df.rdd.map(lambda x: 
    (x["Firstname"] + ", " + x["Lastname"], x["Gender"], x["Salary"] * 2)
    ) 
df2 = rdd2.toDF(["Name", "Gender", "New_Salary"])
df2.show()
```

```
#Referring Column Names
rdd2 = df.rdd.map(lambda x: 
    (x.Firstname + ", " + x.Lastname, x.Gender, x.Salary * 2)
    )
print (type(rdd2))
df2 = rdd2.toDF(["Name", "Gender", "New_Salary"])
df2.show()
```

```
def func1(x):
    row_firstName = x.Firstname
    row_lastName = x.Lastname
    row_name = row_firstName + ", " + row_lastName
    row_gender = x.Gender.lower()
    row_salary = x.Salary * 2
    return (row_name, row_gender, row_salary)

df.rdd.map(lambda x: func1(x)).toDF(["Name", "Gender", "New_Salary"]).show()
df.rdd.map(func1).toDF(["Name", "Gender", "New_Salary"]).show()
```
